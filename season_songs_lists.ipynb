```python
html_source = driver.page_source
soup = bs(html_source, 'html.parser')
```

```python
links = soup.select('body > div:nth-of-type(2) > div:nth-of-type(2) > div > div:nth-of-type(4) > a')

# Lista para almacenar los últimos elementos de los hrefs
x_mas_ids = []

# Iterar sobre todos los enlaces encontrados
for link in links:
    href = link.get('href')
    if href:
        # Split de la URL y tomar el último elemento
        last_element = href.split('/')[-1]
        x_mas_ids.append(last_element)
```

```python
season_songs = []  


for xmas in x_mas_ids:
    season_songs.append({'season': 'xmas', 'id': xmas})  

for halloween in halloween_ids:
    season_songs.append({'season': 'halloween', 'id': halloween})  
season_songs_csv = pd.DataFrame(season_songs)
season_songs_csv.to_csv('season_songs.csv')
```
